```python
## 모듈로딩
import pandas as pd
import numpy as np
import wquantiles as wq
```

```python
## 임의의 데이터
dataA = [70, 72, 75, 75, 76, 80, 82, 85, 88, 89]
dataSR = pd.Series(dataA)
```

```python
## 기술통계값 확인
dataSR.describe(percentiles=[0.3, 0.6, 0.9])
```

```python
## --------------------------------------------------------
## 평균/분산/표준편차 : pandas와 numpy의 분산/표준편차 계산식의 차이로 값 다름
## --------------------------------------------------------
## => pandas
print("평균:", dataSR.mean())
print("분산:", dataSR.var(ddof=0))
print("표준편차:", dataSR.std(ddof=0))

## => numpy
print("평균:", np.mean(dataSR))
print("분산:", np.var(dataSR))
print("표준편차:", np.std(dataSR))
```
