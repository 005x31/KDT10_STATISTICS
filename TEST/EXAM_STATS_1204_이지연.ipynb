```python
## 1) Age와 Fare 각각에 대해 히스토그램과 Boxplot을 그리시오.
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Age
DF.Age.fillna(DF.Age.median(), inplace=True)

axes[0, 0].hist(DF.Age, bins=30)
axes[0, 0].set_title('Age Histogram')

axes[0, 1].boxplot(DF.Age, vert=False)
axes[0, 1].set_title('Age Boxplot')

# Fare
axes[1, 0].hist(DF.Fare, bins=30)
axes[1, 0].set_title('Fare Histogram')

axes[1, 1].boxplot(DF.Fare, vert=False)
axes[1, 1].set_title('Fare Boxplot')

plt.show()
```
